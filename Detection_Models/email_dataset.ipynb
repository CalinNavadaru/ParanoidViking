```python
import pandas as pd

email_df = pd.read_csv("dataset/emails/SpamAssasin.csv")
email_df.dropna(inplace=True)
email_df = email_df[email_df['label'] % 2 == 0]
email_df
```

```python
email_dataset = pd.read_csv("dataset/emails/CEAS_08.csv")

email_dataset = pd.concat([email_dataset, email_df], axis=0, ignore_index=True)

email_dataset.dropna(inplace=True)

email_dataset.drop_duplicates()

email_dataset.to_csv("all_emails3.csv", index=False, encoding='utf-8')

email_dataset
```
